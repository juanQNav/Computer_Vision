```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('test.jpg')
```

```python
plt.imshow(img)
```

```python
h,w,_ = img.shape
rotation_matix = cv2.getRotationMatrix2D((h/2,w/2), angle=45, scale=1.0);
```

```python
print(rotation_matix)
```

```python
rotated_image = cv2.warpAffine(src=img, M=rotation_matix, dsize=(w, h))

plt.imshow(rotated_image)
```
